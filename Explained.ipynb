```python
#step 3 
# TODO 
# 1. Create a function to calculate bond value. (makehams formula)
# 2. We need to have  function to reduce time to maturity by one year
#   2.a For bonds less than 1 year we keep the same
# 3. We need a function to check which yield to price the bonds 

# Step 3 part 1)
def makehams_v1(coup, ytm, Notional, T, freq):
    discount = (1 - 1/( 1+ ytm/freq)**(freq * T))
    result  = Notional * coup/ytm * discount + Notional / (1 + ytm/freq)**(freq*T)
    return(result)


```
