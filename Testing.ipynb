```python
def makehams_formula(coup, rate, time_remaining, freq, Notional,status):
        ye = np.floor(time_remaining)
        # if ye < 0:
        #     ye  += 1 
        adj = time_remaining - ye
        t2 = ye.copy()
        print(adj)
        ## TODO adjust adj  such that we only need to consider the period after coupon.
        def check_adj(adj):
            if adj > 0.5:
                #time_to_maturity = ye + 0.5 
                adj  = (adj- 0.5)/0.5
            else:
                
                adj = adj / 0.5
            return(adj)
        def check_adj_2(adj):
            if adj > 0.5:
                #time_to_maturity = ye + 0.5 
                k =  0.5
            else:
                k = 0
            return(k)
        t1 = adj.copy()
        adjust = adj.apply(lambda x: check_adj(x)) 
        time_to_maturity = ye + t1.apply(lambda x: check_adj_2(x))

        # D is the discounting factor
        D = (1/(1+ rate/freq)**(freq*time_to_maturity))
        # P is the Clean price of the bond at the specific 
        Dirty = (coup /rate * Notional * (1 - D)  + Notional * D) * status
        # Dirty_price represents the Dirty Price of the bond
        P = Dirty * (1 + rate/2)**adjust  -   Notional*coup*adjust
        return(pd.DataFrame.from_dict({"Clean Price":P,"Dirty Price" : Dirty },dtype= np.float64)) 
```

```python
def CDS_pricer(port, status):
    temp = port.copy()
    conditions = [
        (temp["status"] ==  0) & (status == 0), 
     (temp["status"] == 1) | (status == 0)
    ]
    # first condition denotes whether it is  in default or not at the moment we will 
    # change hte code to adjust for teh data
    # We could use a rating system to consider each case but at this moments it is an extra 
    # feature than an actual business need.
    # The second checks if its a bond in default, if true then you also make the recovery
    #temp["value"] = temp.where(cond, port["Notional"]*port["Expected Recovery rate"])
    choices = [temp["Notional"]*temp["Expected Recovery rate"], np.nan]
    temp["value"]= np.select(conditions,choices,default=0)
    return(temp)
```
